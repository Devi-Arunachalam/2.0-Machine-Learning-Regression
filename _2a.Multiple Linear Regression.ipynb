```python
import pandas as pd
dataset=pd.read_csv("50_Startups.csv")

#Perform one hot encoding by using get_dummies function in panda library
dataset=pd.get_dummies(dataset, drop_first=True)

#Get the column names to create input and out put variable
#dataset.columns
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent=dataset[['Profit']]
```

```python
#Split into training set and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
x_train
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
#y=W*x1+b0 for this equation we got value for b1 and bo
regressor.fit(x_train, y_train)

#Viewing the b1 and b0 value
weight=regressor.coef_
#print("Weight of the model={}".format(weight))
print("Weight of the model",weight)
bias=regressor.intercept_
print("Intercept of the model={}".format(bias))
```

```python
y_pred=regressor.predict(x_test)
```

```python
#Evaluate the model
from sklearn.metrics import r2_score
r_score=r2_score(y_test, y_pred)
r_score
```

```python
#Save the model
import pickle
filename="multiLR_final.sav"
pickle.dump(regressor, open(filename,"wb"))
```
